```python
movies=movies.merge(credits,on="title")
movies.head(2)
```

```python
movies=movies[["genres","movie_id","keywords","overview","title","cast","crew"]]
```

```python
#since list indices must be string literals we use ast
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
#we need for cast only top 3 casts
def convertfor3(obj):
    l=[]
    counter=0;
    for i in ast.literal_eval(obj):
        if counter<3:
            l.append(i["name"])
            counter=counter +1;
        else:
            break;
    return l
```

```python
movies["cast"]=movies["cast"].apply(convertfor3)
```

```python
movies
```

```python
#we need only director name from crew
def getdirector(obj):
    l=[];
    for i in  ast.literal_eval(obj):
        if(i['job']=='Director'):
            l.append(i["name"])
    return l
```

```python
movies["crew"]=movies["crew"].apply(getdirector)
```

```python
movies.head()
```

```python
#overview column splitting to list of words
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#since we are doing this for tags a space would make it difficult so removing spaces
movies["genres"]=movies["genres"].apply(lambda x :[i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x :[i.replace(" ","") for i in x])
movies["overview"]=movies["overview"].apply(lambda x :[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x :[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#combining all columns to a single column to get tags
movies["tags"]=movies["genres"]+movies["keywords"]+movies["overview"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
new_df=movies[["title","movie_id","tags"]]
```

```python
new_df.head()
```

```python
#converting tags column to a string
new_df["tags"]=new_df["tags"].apply(lambda x : ' '.join(x))
```

```python
new_df.head()
```

```python
#converting tags column to all lower case
new_df["tags"]=new_df["tags"].apply(lambda x : x.lower())
```

```python
new_df
```

# Stemming

```python
#to remove repeated words in tags like loving,loved,lovable and all as it affects in vectorization stemming is being used
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
new_df.head()
```

we are ready with the dataset to work with now for our requirement

# Vectorization

Why Vectorization? 

converting all tags to a vector in dimensional space so that based on the distances between the vectors similarity can be calculated hence movie can be recommended

so we combine all the tags and count the max say 5000 most frequent words and then check each frequent word count for eqch movie(i.e in tag of that movie) which actually represent a vector

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python

cv=CountVectorizer(max_features=5000,stop_words='english')#no stop words (like are in a etc..)
```

```python
vector=cv.fit_transform(new_df["tags"]).toarray()
vector
```

```python
vector.shape
```

we can see it has been converted to a (4806, 5000) approx 5k dimensional space so each movie is a vector 

means that 4806 movies ka each column of 5k repeated words whether that is present in the movie or not and plotted based on it

Starting Similarity between vectors

we are using cosine distance(angles) not eucledian as dimension increases eucledian is less significant

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

so each movie of 4806 movies ka distance is calculated with the all 4806 movies

```python
similarity[0]#represent movie 1 ka distance with all 4806 movies
```

# Recommending

```python
def recommend(movie):
    movie_indices = new_df[new_df["title"] == movie].index
    if len(movie_indices) == 0:
        print(f"Sorry, '{movie}' is not found in the dataset.")
        return
    
    movie_index = movie_indices[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
#asking input and recommending movies
mov=str(input("Please enter movie name \n"))
print("Recommended movies are \n")
recommend(mov)
```
